```python
import pandas as pd
```

```python
bhNames = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','BLACK','LSTAT','MEDV']

'''
1. CRIM      per capita crime rate by town
    2. ZN        proportion of residential land zoned for lots over 
                 25,000 sq.ft.
    3. INDUS     proportion of non-retail business acres per town
    4. CHAS      Charles River dummy variable (= 1 if tract bounds 
                 river; 0 otherwise)
    5. NOX       nitric oxides concentration (parts per 10 million)
    6. RM        average number of rooms per dwelling
    7. AGE       proportion of owner-occupied units built prior to 1940
    8. DIS       weighted distances to five Boston employment centres
    9. RAD       index of accessibility to radial highways
    10. TAX      full-value property-tax rate per $10,000
    11. PTRATIO  pupil-teacher ratio by town
    12. B        1000(Bk - 0.63)^2 where Bk is the proportion of blacks 
                 by town
    13. LSTAT    % lower status of the population
    14. MEDV     Median value of owner-occupied homes in $1000's
'''
```

```python
df = pd.read_csv('housing.data', delim_whitespace = True, names = bhNames)
```

```python
summaryy = df.describe()
summaryy.transpose()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()

print(scaler.fit(df))

DataScaled = scaler.fit_transform(df)
DataScaled = pd.DataFrame(DataScaled, columns = bhNames)

summary = DataScaled.describe()
summary.transpose()
```

```python
DataScaled
```

```python
import matplotlib.pyplot as plt

# Boxplot
boxplot = DataScaled.boxplot(column=bhNames)
plt.show()

CorData = DataScaled.corr(method='pearson')
# Correlation data
with pd.option_context('display.max_rows', None, 'display.max_columns', CorData.shape[1]):
    print(CorData)
```

```python
from sklearn.model_selection import train_test_split

X = DataScaled.drop('MEDV', axis = 1)
print('X Shape:' ,X.shape)

Y = DataScaled['MEDV']
print('Y Shape:' ,Y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.20, random_state = 5)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras import metrics

Regression = Sequential()
Regression.add(Dense(20, input_dim =13, activation='relu'))
Regression.add(Dense(10, activation='relu'))
Regression.add(Dense(1, activation='linear'))

Regression.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
```

```python
Regression.fit(X_train, y_train, epochs = 1000, verbose = 1)
```

```python
Regression.summary()
```

```python
Y_predictKM = Regression.predict(X_test)
Score = Regression.evaluate(X_test, y_test, verbose = 0)

print("Score: ", Score[0])
```

```python
from sklearn.linear_model import LinearRegression

LModel = LinearRegression()
LModel.fit(X_train, y_train)

PredictedModel = LModel.predict(X_test)

plt.figure(1)
plt.subplot(121)
plt.scatter(y_test, Y_predictKM)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Keras Neural Network Model")

plt.subplot(122)
plt.scatter(y_test, PredictedModel)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Linear Regression Model")

plt.show()

from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, PredictedModel)
print("Mean Squared Error of LinearRegression Model is:" ,mse)
```
